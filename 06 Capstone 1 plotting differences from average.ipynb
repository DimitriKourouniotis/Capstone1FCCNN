#### This Notebook will <li> sort the values of differences from estimated submissions by state <li> draw a lollipop chart of the values

```python
## import packages and set to display charts in document
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Read in the prepared data.

```python
#read in cleaned data comparing proportion
FCC_pop_comments_perc2 =pd.read_csv('FCC_pop_comments_perc2.csv')
```

```python
#explore dataframe
FCC_pop_comments_perc2.head()
```

Order the data and draw the chart highest value to lowest vertically. 

```python
# Order by value, highest to lowest
ordered_df = FCC_pop_comments_perc2.sort_values(by='diff_perc_pop_comments')
state_range = range(1,len(FCC_pop_comments_perc2.index)+1)
 
sns.set()
plt.figure( figsize=(10,10), dpi=80)

#set color, hlines and lollipop head shape
my_color=np.where(ordered_df['diff_perc_pop_comments']<=0, 'orange', 'skyblue')
plt.hlines(y=state_range, xmin=0, xmax=ordered_df['diff_perc_pop_comments'], color=my_color)
plt.plot(ordered_df['diff_perc_pop_comments'], state_range, "o")
 
# Add titles and axis names
plt.yticks(state_range, ordered_df['State_Abbreviation'])
plt.title("Number of Filings by State \nas they differ from Population Distribution")
plt.xlabel('Difference from National Average')
plt.ylabel('State')
```
